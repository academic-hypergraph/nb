```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir as etapas do fluxo de trabalho e suas relaes,
# com quebra de pargrafo para termos com mais de 2 palavras
stages = {
    "Coleta de Dados": [],
    "Pr-processamento\nde Texto": ["Coleta de Dados"],
    "Anlise com\nLLM": ["Pr-processamento\nde Texto"],
    "Extrao de\nInsights": ["Anlise com\nLLM"],
    "Identificao de\nCondies\nNecessrias": ["Extrao de\nInsights"],
    "Modelagem por\nEquaes\nEstruturais": ["Identificao de\nCondies\nNecessrias"],
    "Especificao\ndo Modelo": ["Extrao de\nInsights"],
    "Estimao de\nParmetros": ["Especificao\ndo Modelo"],
    "Avaliao de\nAjuste": ["Estimao de\nParmetros"],
    "Interpretao\nde Resultados": ["Avaliao de\nAjuste"],
    "Publicao\nAcadmica": ["Interpretao\nde Resultados"],
}

# Funo para criar e desenhar o grafo direcionado
def draw_workflow_graph(stages):
    G = nx.DiGraph()

    # Adicionar ns e arestas ao grafo
    for stage, dependencies in stages.items():
        G.add_node(stage)
        for dependency in dependencies:
            G.add_edge(dependency, stage)

    # Definir o layout do grafo com espaamento
    pos = nx.spring_layout(G, seed=62, k=3.3)

    # Definir cores pastel escuras (prximas ao preto) para os ns
    color_map = {
        "Coleta de Dados": "#4B3C4D",
        "Pr-processamento\nde Texto": "#4B3C4D",
        "Anlise com\nLLM": "#3C4B5A",
        "Extrao de\nInsights": "#3C4B5A",
        "Identificao de\nCondies\nNecessrias": "#5A3C4B",
        "Modelagem por\nEquaes\nEstruturais": "#5A3C4B",
        "Especificao\ndo Modelo": "#3C5A4B",
        "Estimao de\nParmetros": "#3C5A4B",
        "Avaliao de\nAjuste": "#3C5A4B",
        "Interpretao\nde Resultados": "#4B5A3C",
        "Publicao\nAcadmica": "#4B5A3C",
    }

    # Desenhar os ns com estilo e cores
    nx.draw_networkx_nodes(G, pos, node_size=2900, node_color=[color_map[node] for node in G.nodes()],
                           linewidths=2, edgecolors="black", node_shape='o')

    # Desenhar as arestas direcionadas com estilo
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="darkgray",
                           arrowsize=12, arrowstyle='->', connectionstyle='arc3,rad=0.18')

    # Adicionar rtulos aos ns com estilo
    nx.draw_networkx_labels(G, pos, font_size=6, font_family="sans", font_weight="bold",
                            font_color="white")

    # Adicionar ttulo e remover eixos
    plt.title("Fluxo de Trabalho Acadmico: Anlise com LLMs, NCA e SEM", fontsize=12, fontweight="bold", color="white")
    plt.axis("off")

    # Definir fundo preto
    plt.gcf().set_facecolor("black")

    # Ajustar o tamanho da figura
    plt.figure(figsize=(36, 16))

    # Mostrar a visualizao
    plt.show()

# Executar a funo para desenhar o grafo
if __name__ == "__main__":
    draw_workflow_graph(stages)
```
