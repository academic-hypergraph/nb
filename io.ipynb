```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir as etapas do fluxo de trabalho e suas relações,
# com quebra de parágrafo para termos com mais de 2 palavras
stages = {
    "Coleta de Dados": [],
    "Pré-processamento\nde Texto": ["Coleta de Dados"],
    "Análise com\nLLM": ["Pré-processamento\nde Texto"],
    "Extração de\nInsights": ["Análise com\nLLM"],
    "Identificação de\nCondições\nNecessárias": ["Extração de\nInsights"],
    "Modelagem por\nEquações\nEstruturais": ["Identificação de\nCondições\nNecessárias"],
    "Especificação\ndo Modelo": ["Extração de\nInsights"],
    "Estimação de\nParâmetros": ["Especificação\ndo Modelo"],
    "Avaliação de\nAjuste": ["Estimação de\nParâmetros"],
    "Interpretação\nde Resultados": ["Avaliação de\nAjuste"],
    "Publicação\nAcadêmica": ["Interpretação\nde Resultados"],
}

# Função para criar e desenhar o grafo direcionado
def draw_workflow_graph(stages):
    G = nx.DiGraph()

    # Adicionar nós e arestas ao grafo
    for stage, dependencies in stages.items():
        G.add_node(stage)
        for dependency in dependencies:
            G.add_edge(dependency, stage)

    # Definir o layout do grafo com espaçamento
    pos = nx.spring_layout(G, seed=62, k=3.3)

    # Definir cores pastel escuras (próximas ao preto) para os nós
    color_map = {
        "Coleta de Dados": "#4B3C4D",
        "Pré-processamento\nde Texto": "#4B3C4D",
        "Análise com\nLLM": "#3C4B5A",
        "Extração de\nInsights": "#3C4B5A",
        "Identificação de\nCondições\nNecessárias": "#5A3C4B",
        "Modelagem por\nEquações\nEstruturais": "#5A3C4B",
        "Especificação\ndo Modelo": "#3C5A4B",
        "Estimação de\nParâmetros": "#3C5A4B",
        "Avaliação de\nAjuste": "#3C5A4B",
        "Interpretação\nde Resultados": "#4B5A3C",
        "Publicação\nAcadêmica": "#4B5A3C",
    }

    # Desenhar os nós com estilo e cores
    nx.draw_networkx_nodes(G, pos, node_size=2900, node_color=[color_map[node] for node in G.nodes()],
                           linewidths=2, edgecolors="black", node_shape='o')

    # Desenhar as arestas direcionadas com estilo
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="darkgray",
                           arrowsize=12, arrowstyle='->', connectionstyle='arc3,rad=0.18')

    # Adicionar rótulos aos nós com estilo
    nx.draw_networkx_labels(G, pos, font_size=6, font_family="sans", font_weight="bold",
                            font_color="white")

    # Adicionar título e remover eixos
    plt.title("Fluxo de Trabalho Acadêmico: Análise com LLMs, NCA e SEM", fontsize=12, fontweight="bold", color="white")
    plt.axis("off")

    # Definir fundo preto
    plt.gcf().set_facecolor("black")

    # Ajustar o tamanho da figura
    plt.figure(figsize=(36, 16))

    # Mostrar a visualização
    plt.show()

# Executar a função para desenhar o grafo
if __name__ == "__main__":
    draw_workflow_graph(stages)
```
