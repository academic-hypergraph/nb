```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir as hiperarestas do hipergrafo para NCA e SEM
hyperedges = [
    ("Coleta de Dados", "Identificao de Condies Necessrias", "NCA"),
    ("NCA", "Especificao de Modelo", "SEM"),
    ("Especificao de Modelo", "Avaliao de Ajuste", "SEM"),
    ("Avaliao de Ajuste", "Interpretao de Resultados", "Publicao Acadmica"),
    # Adicione outras etapas conforme necessrio
]

# Funo para criar e desenhar o hipergrafo direcionado
def draw_academic_hypergraph(hyperedges):
    G = nx.DiGraph()

    # Extrair ns nicos das hiperarestas
    nodes = set()
    for edge in hyperedges:
        nodes.update(edge)

    # Adicionar ns ao grafo
    G.add_nodes_from(nodes)

    # Adicionar hiperarestas direcionadas ao grafo
    for edge in hyperedges:
        source = edge[0]
        for target in edge[1:]:
            G.add_edge(source, target)

    # Definir o layout do grafo com espaamento acadmico
    pos = nx.spring_layout(G, seed=42, k=0.8)

    # Desenhar os ns
    nx.draw_networkx_nodes(G, pos, node_size=1500, node_color="lightblue")

    # Desenhar as arestas direcionadas
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=1, edge_color="gray", arrowsize=20)

    # Adicionar rtulos aos ns
    nx.draw_networkx_labels(G, pos, font_size=9, font_family="sans-serif")

    # Ajustar o tamanho da figura
    plt.figure(figsize=(12.8, 7.2))

# Executar a funo para desenhar o hipergrafo
if __name__ == "__main__":
    draw_academic_hypergraph(hyperedges)
```
